```python
import numpy as np
import pandas as pd

df_raw = pd.read_csv(
    "out.txt", 
    delim_whitespace=True, 
    names=["World index", "Success", "Hit Obstacle", "Timeout", "Time", "Score"])

success_rate = df_raw["Success"].mean()

df_success = df_raw[df_raw["Success"] == 1]

average_time = df_success["Time"].mean()
average_score = df_success["Score"].mean()

print(f"Success rate : {success_rate:.2f}\nAverage Time : {average_time:.2f}\nAverage Score : {average_score:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot(title, x, y, y_max, y_min=0):
    plt.clf()
    sns.set_style("darkgrid")
    sns.set_context("poster")
    plt.figure(figsize=(10, 6))
    plt.title(title)
    plt.axis([-1, len(y), y_min, y_max])
    plt.bar(x, y, width=0.5)
    for _x, _y in zip(range(len(x)), y):
        plt.text(_x, _y-(_y*0.05), f"{_y:.4f}%", fontsize="small", color="black", ha="center", va="center")
    plt.show()

# All Average
# Time, Metric, Success, Collision, Timeout
local_planner_names = ["dwa", "eband", "teb"]
data = [
    [27.1010, 0.2170, 1.0000, 0.0000, 0.0000],
    [16.6017, 0.2273, 0.9091, 0.0909, 0.0000],   
    [25.4671, 0.2017, 0.9091, 0.0909, 0.0000]
]
data = np.array(data)

plot_titles = ["Avg Time", "Avg Metric", "Avg Success", "Avg Collision", "Avg Timeout"]
plot_y_maxs = [100.0, 1.0, 1.0, 1.0, 1.0]
for index, title in enumerate(plot_titles):
    plot(title, local_planner_names, data[:, index], y_max=plot_y_maxs[index])
```
